```
import pandas as pd
```

```
data = [0.5, None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
s = pd.Series(data)
```

```
s
```

```
s.fillna(0)
```

```
s.fillna(method = 'ffill') #substitui o valor faltante com o último valor válido (de cima pra baixo)
```

```
s.fillna(method = 'bfill') # substitui o valor faltante com o último valor válido (de baixo pra cima)
```

```
s.fillna(s.mean()) #substitui os valores nulos pela média dos valores não nulos
```

```
s.fillna(method = 'ffill', limit = 1)
```

```
s1 = s.fillna(method = 'ffill', limit = 1)
s1.fillna(method = 'bfill', limit = 1)
```
